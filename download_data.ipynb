```python

import pandas_datareader.data as web
import datetime as dt

def download_nasdaq():
    start = dt.datetime(2010, 1, 1)
    end = dt.datetime(2024, 12, 31)

    print("Downloading Nasdaq-100 data from Stooq...")
    df = web.DataReader("^NDX", "stooq", start, end)

    if df.empty:
        raise RuntimeError("Download failed: received empty dataset")

    # Stooq renvoie les dates Ã  l'envers
    df = df.sort_index()

    df.to_csv("data/raw/nasdaq100_prices.csv")
    print("Market data successfully saved to data/raw/nasdaq100_prices.csv")

if __name__ == "__main__":
    download_nasdaq()
```
